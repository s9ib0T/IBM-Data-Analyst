# **Collecting Job Data Using API**


### Objectives


*   Collect job data from Jobs API
*   Store the collected data into an excel spreadsheet.


```python
# Requirements
import requests
api_url="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"
response = requests.get(api_url)
jobs_data = response.json()
```

### 1. Write a function to get the number of jobs for the Python technology

```python
def count_jobs_technology(technology):
    job_count = 0
    
    for job in jobs_data:
        key_skills = job.get("Key Skills")
        if technology in key_skills:
            job_count += 1
    
    return job_count
```

```python
python_jobs = count_jobs_technology('Python')
print(f"Number of jobs for Python technology: {python_jobs}")
```

### 2. Write a function to find number of jobs in US for a location


```python
def count_jobs_Location(location):
    job_count = 0
    
    for job in jobs_data:
        city = job.get('Location')
        if city == location:
            job_count += 1
   
    return location, job_count
```

```python
some_cities = ["Los Angeles", "New York", "San Francisco", "Washington DC", "Seattle"] # Example cities
for city in some_cities:
    location, job_count = count_jobs_Location(city) 
    print(f"Number of jobs for location {location}: {job_count}")
```

### 3.  Store the results in an excel file


#### 3.1 Collect the number of job postings and write the results in an excel spreadsheet for the following languages:

*   C
*   C#
*   C++
*   Java
*   JavaScript
*   Python
*   Scala
*   Oracle
*   SQL Server
*   MySQL Server
*   PostgreSQL
*   MongoDB

```python
data_technology = []
technologies = [
    "C", "C#", "C++", "Java", "JavaScript", "Python",
    "Scala", "Oracle", "SQL Server", "MySQL Server",
    "PostgreSQL", "MongoDB"
]
for technology in technologies:
    data_technology.append((technology, count_jobs_technology(technology)))
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(data_technology, columns=["Technology", "Job Postings"])
print(df1)
```

Convert dataframe to excel

```python
file_path = "job-postings.xlsx"
df1.to_excel(file_path, index=False, sheet_name="Jobs per Technology")
```

#### 3.2 Collect the number of job postings and write the results in an excel spreadsheet for the locations:


* Los Angeles
* New York
* San Francisco
* Washington DC
* Seattle
* Austin
* Detroit

```python
data_location = []
locations = [
    "Los Angeles",
    "New York",
    "San Francisco",
    "Washington DC",
    "Seattle",
    "Austin",
    "Detroit"
]
for location in locations:
    data_location.append(count_jobs_Location(location))
```

```python
df2 = pd.DataFrame(data_location, columns=['Location', 'Job Postings'])
print(df2)
```

Write to the previously created excel file

```python
with pd.ExcelWriter(file_path, engine="openpyxl", mode="a") as writer:
    df2.to_excel(writer, index=False, sheet_name="Jobs per Location")
```
